```python
# hapus node
class Node:
  def __init__(self, dataval=None):  # Changed 'dataval' to 'data'
    self.dataval = dataval
    self.nextval = None

class LinkedList:
  def __init__(self):
    self.headval = None

# Print the linked list
  def listprint(self):
    printval = self.headval
    while printval  is not None:
      print (printval.dataval)
      printval = printval.nextval

# Function to add node in the begining
  def AddBegining(self, newdata):
    NewNode = Node(newdata)

# Update the new nodes next val to exxiting node
    NewNode.nextval = self.headval
    self.headval = NewNode

# Fuction to remove node by key
  def RemoveNode(self, Removekey): # Added 'Removekey' parameter
     HeadVal = self.headval
     # Case 1: Head node itself holds the key to be deleted
     if (HeadVal is not None):
        if (HeadVal.dataval == Removekey):
           self.headval = HeadVal.nextval
           return
     # Search for the key to be deleted, keep track of the previous node
     prev = None
     while (HeadVal is not None and HeadVal.dataval != Removekey):
        prev = HeadVal
        HeadVal = HeadVal.nextval
     # If key was not present in linked list
     if (HeadVal == None):
        return
     # Unlink the node from linked list
     if prev is not None:
        prev.nextval = HeadVal.nextval

# create linked list named Li
Li = LinkedList()

#add node
Li.AddBegining("April")
Li.AddBegining("Maret")
Li.AddBegining("Febuari")
print("Sebelum dilakukan remove")
Li.listprint()
print("==================================================")

#remove node by key = "Maret"
Li.RemoveNode("Maret")
print("Setelah dilakukan remove dengan kunci 'Maret'")
Li.listprint()
```

```python
# Hapus di akhir
class Node:
  def __init__(self, dataval=None):
    self.dataval = dataval
    self.nextval = None

class LinkedList:
  def __init__(self):
    self.headval = None

# Print the linked list
  def listprint(self):
    printval = self.headval
    while printval  is not None:
      print (printval.dataval)
      printval = printval.nextval

# Function to add node to the end
  def AddInEnd(self, newdata):
     NewNode = Node(newdata)
     if self.headval is None:
        self.headval = NewNode
        return
     last = self.headval
     while(last.nextval):
        last = last.nextval
     last.nextval=NewNode

# Fuction to remove last node
  def RemoveEnd(self):
    last = self.headval
    while(last.nextval):
      if last.nextval == None:
         break
      prev = last
      last = last.nextval
    if (last == None):
      return
    prev.nextval = last.nextval
    last = None

# create linked list named Li
Li = LinkedList()

# add node
Li.AddInEnd("April")
Li.AddInEnd("Maret")
Li.AddInEnd("Febuari")
print("Sebelum dilakukan remove")
Li.listprint()
print("==================================================")

#remove last node
Li.RemoveEnd()
print("Setelah dilakukan remove node terakhir")
Li.listprint()
```

```python
# Tugas no 1
# ==============================
# DOUBLE LINKED LIST
# ==============================

class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None
        self.prevval = None


class DoubleLinkedList:
    def __init__(self):
        self.headval = None

    # Cetak isi list
    def listprint(self):
        printval = self.headval
        while printval is not None:
            print(printval.dataval, end=" <-> ")
            printval = printval.nextval
        print("None")

    # Sisip di tengah (setelah key tertentu)
    def AddInBetween(self, key, newdata):
        curr = self.headval

        while curr is not None:
            if curr.dataval == key:
                break
            curr = curr.nextval

        if curr is None:
            print("Node tidak ditemukan")
            return

        NewNode = Node(newdata)

        NewNode.nextval = curr.nextval
        NewNode.prevval = curr

        if curr.nextval is not None:
            curr.nextval.prevval = NewNode

        curr.nextval = NewNode

    # Hapus berdasarkan kata kunci
    def RemoveByKey(self, key):
        curr = self.headval

        if curr is not None and curr.dataval == key:
            self.headval = curr.nextval
            if self.headval is not None:
                self.headval.prevval = None
            return

        while curr is not None:
            if curr.dataval == key:
                break
            curr = curr.nextval

        if curr is None:
            print("Data tidak ditemukan")
            return

        if curr.nextval is not None:
            curr.nextval.prevval = curr.prevval

        if curr.prevval is not None:
            curr.prevval.nextval = curr.nextval


# ==============================
# EKSEKUSI PROGRAM
# ==============================

DLL = DoubleLinkedList()

# Buat 3 node
n1 = Node("Senin")
n2 = Node("Selasa")
n3 = Node("Rabu")

# Hubungkan node
DLL.headval = n1
n1.nextval = n2
n2.prevval = n1
n2.nextval = n3
n3.prevval = n2

print("Data awal:")
DLL.listprint()

# Sisip tengah
DLL.AddInBetween("Senin", "Libur")
print("\nSetelah sisip tengah:")
DLL.listprint()

# Hapus dengan kata kunci
DLL.RemoveByKey("Selasa")
print("\nSetelah hapus kata kunci 'Selasa':")
DLL.listprint()
```

```python
#Tuga no 2
# ==============================
# CIRCULAR LINKED LIST
# ==============================

class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None


class CircularLinkedList:
    def __init__(self):
        self.headval = None

    # Cetak isi circular list
    def listprint(self):
        temp = self.headval
        if temp is None:
            print("List kosong")
            return

        while True:
            print(temp.dataval, end=" -> ")
            temp = temp.nextval
            if temp == self.headval:
                break
        print("(Balik ke Head)")

    # Sisip di akhir
    def AddInEnd(self, newdata):
        NewNode = Node(newdata)

        if self.headval is None:
            self.headval = NewNode
            NewNode.nextval = self.headval
            return

        last = self.headval
        while last.nextval != self.headval:
            last = last.nextval

        last.nextval = NewNode
        NewNode.nextval = self.headval

    # Hapus di akhir
    def RemoveEnd(self):
        if self.headval is None:
            return

        if self.headval.nextval == self.headval:
            self.headval = None
            return

        temp = self.headval
        while temp.nextval.nextval != self.headval:
            temp = temp.nextval

        temp.nextval = self.headval


# ==============================
# EKSEKUSI PROGRAM
# ==============================

CLL = CircularLinkedList()

# Buat 3 node
n1 = Node("Jam 1")
n2 = Node("Jam 2")
n3 = Node("Jam 3")

# Hubungkan circular
CLL.headval = n1
n1.nextval = n2
n2.nextval = n3
n3.nextval = n1

print("Data awal:")
CLL.listprint()

# Sisip akhir
CLL.AddInEnd("Jam 4")
print("\nSetelah sisip akhir:")
CLL.listprint()

# Hapus akhir
CLL.RemoveEnd()
print("\nSetelah hapus akhir:")
CLL.listprint()
```
